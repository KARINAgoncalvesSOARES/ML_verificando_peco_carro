```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

##Importando a base de dados:

```python
df_videos = pd.read_excel('/content/drive/MyDrive/lira#/Cópia de videosYT.xlsx')
display(df_videos)
```

##Tipos de Gáficos

Agora vamos para a parte de tipos de gráficos onde temos a separação desses gráficos por categoria.

Documentação : https://seaborn.pydata.org/api.html#distribution-api image.png

image.png

##Criando um gráfico de dispersão:

Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df_videos,
                      x="Nº de Views",
                      y="Nº de Likes",
                      hue="Responsável",
                      style="Responsável",
                      palette=["red","black"],
                      size="Nº Inscrições",
                      legend="full")
```

##Criando um gráfico de dispersão(relacional):
Gráfico que irá fazer relação entre as informações que temos.

Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
fig = sns.relplot(data=df_videos, x="Nº de Views",y="Nº de Likes",col="Categoria")#kind="line"
fig.set_axis_labels(x_var="N° views", y_var="N° Likes")
fig.set_titles("Categoria:{col_name}")
plt.show(fig)
```

##Criando um gráfico de linha:
Documentação : https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
df_inscritos = pd.read_excel('/content/drive/MyDrive/lira#/Cópia de videosYT.xlsx', "Inscritos")
display(df_inscritos)

graf_linha = sns.lineplot(data=df_inscritos, x="Mês/Ano", y="Inscritos", color="red")
graf_linha.set_title("Crescimento Inscritos")
plt.show()
```

##Criando Histogramas:
Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
fig = sns.displot(data = df_videos,
                  x="Nº de Likes",
                  hue="Responsável", #hue = significa que o gráfico será dividido em cores diferentes com base nos valores únicos da coluna "Responsável".
                  col="Categoria") #Cada subgráfico mostrará a distribuição dos likes para aquela categoria específica.
plt.show()
```

```python
# O gráfico de pares mostra a relação entre todas as combinações possíveis de variáveis numéricas no DataFrame

fig = sns.pairplot(data=df_videos, hue="Categoria")
plt.show()
```

```python
fig = sns.regplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes") 
plt.show(fig)
```

```python
fig = sns.lmplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes", hue= "Responsável", markers = ['o','x']) 
plt.show(fig)
```
